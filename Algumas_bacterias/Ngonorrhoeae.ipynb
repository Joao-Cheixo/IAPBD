```python
#importação dos diferentes packages

from Bio import SeqIO
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
from Bio import SeqFeature
```

```python
#ler o ficheiro

import re

fonte = open("Ngonorrhoeae.gb",'r')
Ngonorrhoeae = fonte.read()
fonte.close()
print( Ngonorrhoeae )
```

```python
#imprimir o link NCBI

existe = re.search(r'ACCESSION\s+[^\s]+', Ngonorrhoeae)
if existe:
    m = re.match(r'ACCESSION\s+([^\s]+)', existe[0] )
    if m:
        id = m.group(1)
        print( "https://www.ncbi.nlm.nih.gov/nuccore/{}".format( id ) )
else:
    print( "Padrão não encontrado" )
```

```python
#imprimir o título

existe = re.findall(r'TITLE\s+.*?(?=JOURNAL)', Ngonorrhoeae , re.DOTALL)
if existe:
    for title in existe:
        m = re.match( r'TITLE\s+(.+)', title, re.DOTALL )
        print( re.sub(r'\s+', ' ', m.group(1) ) )
```

```python
#imprimir os autores

existe = re.findall(r'AUTHORS\s+.*?(?=JOURNAL)', Ngonorrhoeae, re.DOTALL)
if existe:
    for authors in existe:
        m = re.match( r'AUTHORS\s+(.+)', authors, re.DOTALL )
        print( re.sub(r'\s+', ' ', m.group(1) ) )
```

```python
#imprimir o locus e a source

flag = False
source = ""
locusid = ""
fonte = open("Ngonorrhoeae.gb",'r')
for linha in fonte:
    if re.search(r"^LOCUS", linha):
        s = re.match(r"LOCUS\s+([^ ]+)", linha)
        if s:
            locusid = s.group(1)    
    if re.search(r"^SOURCE", linha):
        s = re.match(r"SOURCE\s+(.+)", linha)
        if s:
            source = s.group(1)
    if re.search(r"//", linha):
        flag = False  
    if re.search(r"ORIGIN", linha):
        flag = True
fonte.close()
print( locusid, source)
```

```python
#ler o ficheiro
Ngonorrhoeae = SeqIO.read("Ngonorrhoeae.gb", "genbank")

#Bactéria
print("ID: ", Ngonorrhoeae.id)
print("Descrição: ", Ngonorrhoeae.description)

#Locus
print("Locus: ", Ngonorrhoeae.name)
print("bp: ", len(Ngonorrhoeae))

#Sequência
print("Tamanho da sequência: ", len(Ngonorrhoeae), "pb")

#Features
print("Features: ", Ngonorrhoeae.features)
print("Tamanho das features: ", len(Ngonorrhoeae.features))
#Identificação de features do tipo CDS e a sua localização:
featcds = [] 
for i in range(len(Ngonorrhoeae.features)):
    if Ngonorrhoeae.features[i].type=='CDS':
        featcds.append(i)
for k in featcds:
    print("Features do tipo CDS e a sua localização:\n",Ngonorrhoeae.features[k].location)
    print("tamanho das features do tipo CDS: ",len(featcds))

featgenes = []  #número de genes anotados no registo (feature tipo “gene”)
for i in range(len(Ngonorrhoeae.features)):
    if Ngonorrhoeae.features[i].type == "gene":
        featgenes.append(i)
print("tamanho das features do tipo gene: ", len(featgenes))

#Reference
print("Referências bibliográficas relacionadas com registo:\n", Ngonorrhoeae.annotations['references'])

#Anotations
print("Tamanho das annotations:", len(Ngonorrhoeae.annotations) )
print("from:", Ngonorrhoeae.annotations["source"])
print(Ngonorrhoeae.letter_annotations)
print("Organismo:", Ngonorrhoeae.annotations["organism"])
print("Taxonomia:", Ngonorrhoeae.annotations["taxonomy"])
print("Tipo de molécula: ", Ngonorrhoeae.annotations['molecule_type'])
print("Código do registo: ", Ngonorrhoeae.annotations['accessions'])
print ("Keywords:\n",Ngonorrhoeae.annotations.keys())


print(Ngonorrhoeae.dbxrefs)
```
