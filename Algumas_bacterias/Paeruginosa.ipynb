```python
#importação dos diferentes packages

from Bio import SeqIO
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
from Bio import SeqFeature
```

```python
#ler o ficheiro

import re

fonte = open("Paeruginosa.gb",'r')
Paeruginosa = fonte.read()
fonte.close()
print( Paeruginosa )
```

```python
#imprimir o link NCBI

existe = re.search(r'ACCESSION\s+[^\s]+', Paeruginosa)
if existe:
    m = re.match(r'ACCESSION\s+([^\s]+)', existe[0] )
    if m:
        id = m.group(1)
        print( "https://www.ncbi.nlm.nih.gov/nuccore/{}".format( id ) )
else:
    print( "Padrão não encontrado" )
```

```python
#imprimir o título

existe = re.findall(r'TITLE\s+.*?(?=JOURNAL)', Paeruginosa , re.DOTALL)
if existe:
    for title in existe:
        m = re.match( r'TITLE\s+(.+)', title, re.DOTALL )
        print( re.sub(r'\s+', ' ', m.group(1) ) )
```

```python
existe = re.findall(r'AUTHORS\s+.*?(?=JOURNAL)', Paeruginosa, re.DOTALL)
if existe:
    for authors in existe:
        m = re.match( r'AUTHORS\s+(.+)', authors, re.DOTALL )
        print( re.sub(r'\s+', ' ', m.group(1) ) )
```

```python
#imprimir o locus e a source

flag = False
source = ""
locusid = ""
fonte = open("Paeruginosa.gb",'r')
for linha in fonte:
    if re.search(r"^LOCUS", linha):
        s = re.match(r"LOCUS\s+([^ ]+)", linha)
        if s:
            locusid = s.group(1)    
    if re.search(r"^SOURCE", linha):
        s = re.match(r"SOURCE\s+(.+)", linha)
        if s:
            source = s.group(1)
    if re.search(r"//", linha):
        flag = False  
    if re.search(r"ORIGIN", linha):
        flag = True
fonte.close()
print( locusid, source)
```

```python
#ler o ficheiro
Paeruginosa = SeqIO.read("Paeruginosa.gb", "genbank")

#Bactéria
print("ID: ", Paeruginosa.id)
print("Descrição: ", Paeruginosa.description)

#Locus
print("Locus: ", Paeruginosa.name)
print("bp: ", len(Paeruginosa))

#Sequência
print("Tamanho da sequência: ", len(Paeruginosa), "pb")

#Reference
print("Referências bibliográficas relacionadas com registo:\n", Paeruginosa.annotations['references'])

#Annotations
print("Tamanho das annotations:", len(Paeruginosa.annotations) )
print("from:", Paeruginosa.annotations["source"])
print(Paeruginosa.letter_annotations)
print("Organismo:", Paeruginosa.annotations["organism"])
print("Taxonomia:", Paeruginosa.annotations["taxonomy"])
print("Tipo de molécula: ", Paeruginosa.annotations['molecule_type'])
print("Código do registo: ", Paeruginosa.annotations['accessions'])
print ("Keywords:\n",Paeruginosa.annotations.keys())


print(Paeruginosa.dbxrefs)
```

```python
#Features
print("Features: ", Paeruginosa.features)
print("Tamanho das features: ", len(Paeruginosa.features))
#Identificação de features do tipo CDS e a sua localização:
featcds = [] 
for i in range(len(Paeruginosa.features)):
    if Paeruginosa.features[i].type=='CDS':
        featcds.append(i)
for k in featcds:
    print("Features do tipo CDS e a sua localização:\n",Paeruginosa.features[k].location)
    print("tamanho das features do tipo CDS: ",len(featcds))

featgenes = []  #número de genes anotados no registo (feature tipo “gene”)
for i in range(len(Paeruginosa.features)):
    if Paeruginosa.features[i].type == "gene":
        featgenes.append(i)
print("tamanho das features do tipo gene: ", len(featgenes))
```
