```python
#importação dos diferentes packages

from Bio import SeqIO
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
from Bio import SeqFeature
```

```python
#ler o ficheiro

import re

fonte = open("Saureus.gb",'r')
Saureus = fonte.read()
fonte.close()
print( Saureus )
```

```python
#imprimir o link NCBI

existe = re.search(r'ACCESSION\s+[^\s]+', Saureus)
if existe:
    m = re.match(r'ACCESSION\s+([^\s]+)', existe[0] )
    if m:
        id = m.group(1)
        print( "https://www.ncbi.nlm.nih.gov/nuccore/{}".format( id ) )
else:
    print( "Padrão não encontrado" )
```

```python
#imprimir o título

existe = re.findall(r'TITLE\s+.*?(?=JOURNAL)', Saureus, re.DOTALL)
if existe:
    for title in existe:
        m = re.match( r'TITLE\s+(.+)', title, re.DOTALL )
        print( re.sub(r'\s+', ' ', m.group(1) ) )
```

```python
#imprimir o locus e a source

flag = False
source = ""
locusid = ""
fonte = open("Saureus.gb",'r')
for linha in fonte:
    if re.search(r"^LOCUS", linha):
        s = re.match(r"LOCUS\s+([^ ]+)", linha)
        if s:
            locusid = s.group(1)    
    if re.search(r"^SOURCE", linha):
        s = re.match(r"SOURCE\s+(.+)", linha)
        if s:
            source = s.group(1)
    if re.search(r"//", linha):
        flag = False  
    if re.search(r"ORIGIN", linha):
        flag = True
fonte.close()
print( locusid, source)
```

```python
#ler o ficheiro
Saureus = SeqIO.read("Saureus.gb", "genbank")

#Bactéria
print("ID: ", Saureus.id)
print("Descrição: ", Saureus.description)

#Locus
print("Locus: ", Saureus.name)
print("bp: ", len(Saureus))

#Sequência
print("Tamanho da sequência: ", len(Saureus), "pb")

#Features
print("Features: ", Saureus.features)
print("Tamanho das features: ", len(Saureus.features))
#Identificação de features do tipo CDS e a sua localização:
featcds = [] 
for i in range(len(Saureus.features)):
    if Saureus.features[i].type=='CDS':
        featcds.append(i)
for k in featcds:
    print("Features do tipo CDS e a sua localização:\n",Saureus.features[k].location)
    print(len(featcds))

featgenes = []  #número de genes anotados no registo (feature tipo “gene”)
for i in range(len(Saureus.features)):
    if Saureus.features[i].type == "gene":
        featgenes.append(i)
print(len(featgenes))

#Reference
print("Referências bibliográficas relacionadas com registo:\n", Saureus.annotations['references'])

#Anotations
print(len(Saureus.annotations) )
print("from:", Saureus.annotations["source"])
print(Saureus.letter_annotations)
print("Organismo:", Saureus.annotations["organism"])
print("Taxonomia:", Saureus.annotations["taxonomy"])
print("Tipo de molécula: ", Saureus.annotations['molecule_type'])
print("Código do registo: ", Saureus.annotations['accessions'])
print ("Keywords:\n",Saureus.annotations.keys())


print(Saureus.dbxrefs)
```
