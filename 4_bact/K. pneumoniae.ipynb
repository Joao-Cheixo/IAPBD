```python
#importação dos diferentes packages

from Bio import SeqIO
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
from Bio import SeqFeature
```

```python
#ler o ficheiro

import re

fonte = open("Kpneumoniae.gb",'r')
Kpneumoniae = fonte.read()
fonte.close()
print( Kpneumoniae )
```

```python
#imprimir o link NCBI

existe = re.search(r'ACCESSION\s+[^\s]+', Kpneumoniae)
if existe:
    m = re.match(r'ACCESSION\s+([^\s]+)', existe[0] )
    if m:
        id = m.group(1)
        print( "https://www.ncbi.nlm.nih.gov/nuccore/{}".format( id ) )
else:
    print( "Padrão não encontrado" )
```

```python
#imprimir o título

existe = re.findall(r'TITLE\s+.*?(?=JOURNAL)', Kpneumoniae, re.DOTALL)
if existe:
    for title in existe:
        m = re.match( r'TITLE\s+(.+)', title, re.DOTALL )
        print( re.sub(r'\s+', ' ', m.group(1) ) )
```

```python
#imprimir o locus e a source

flag = False
source = ""
locusid = ""
fonte = open("Kpneumoniae.gb",'r')
for linha in fonte:
    if re.search(r"^LOCUS", linha):
        s = re.match(r"LOCUS\s+([^ ]+)", linha)
        if s:
            locusid = s.group(1)    
    if re.search(r"^SOURCE", linha):
        s = re.match(r"SOURCE\s+(.+)", linha)
        if s:
            source = s.group(1)
    if re.search(r"//", linha):
        flag = False  
    if re.search(r"ORIGIN", linha):
        flag = True
fonte.close()
print( locusid, source)
```

```python
#ler o ficheiro
Kpneumoniae = SeqIO.read("Kpneumoniae.gb", "genbank")

#Bactéria
print("ID: ", Kpneumoniae.id)
print("Descrição: ", Kpneumoniae.description)

#Locus
print("Locus: ", Kpneumoniae.name)
print("bp: ", len(Kpneumoniae))

#Sequência
print("Tamanho da sequência: ", len(Kpneumoniae), "pb")

#Features
print("Features: ", Kpneumoniae.features)
print("Tamanho das features: ", len(Kpneumoniae.features))
#Identificação de features do tipo CDS e a sua localização:
featcds = [] 
for i in range(len(Kpneumoniae.features)):
    if Kpneumoniae.features[i].type=='CDS':
        featcds.append(i)
for k in featcds:
    print("Features do tipo CDS e a sua localização:\n",Kpneumoniae.features[k].location)
    print(len(featcds))

featgenes = []  #número de genes anotados no registo (feature tipo “gene”)
for i in range(len(Kpneumoniae.features)):
    if Kpneumoniae.features[i].type == "gene":
        featgenes.append(i)
print(len(featgenes))

#Reference
print("Referências bibliográficas relacionadas com registo:\n", Kpneumoniae.annotations['references'])

#Anotations
print(len(Kpneumoniae.annotations) )
print("from:", Kpneumoniae.annotations["source"])
print(Kpneumoniae.letter_annotations)
print("Organismo:", Kpneumoniae.annotations["organism"])
print("Taxonomia:", Kpneumoniae.annotations["taxonomy"])
print("Tipo de molécula: ", Kpneumoniae.annotations['molecule_type'])
print("Código do registo: ", Kpneumoniae.annotations['accessions'])
print ("Keywords:\n",Kpneumoniae.annotations.keys())


print(Kpneumoniae.dbxrefs)
```
