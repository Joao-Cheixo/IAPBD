```python
#importação dos diferentes packages

from Bio import SeqIO
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
from Bio import SeqFeature
```

```python
#ler o ficheiro

import re

fonte = open("Mtuberculosis.gb",'r')
Mtuberculosis = fonte.read()
fonte.close()
print( Mtuberculosis )
```

```python
#imprimir o link NCBI

existe = re.search(r'ACCESSION\s+[^\s]+', Mtuberculosis)
if existe:
    m = re.match(r'ACCESSION\s+([^\s]+)', existe[0] )
    if m:
        id = m.group(1)
        print( "https://www.ncbi.nlm.nih.gov/nuccore/{}".format( id ) )
else:
    print( "Padrão não encontrado" )
```

```python
#imprimir o título

existe = re.findall(r'TITLE\s+.*?(?=JOURNAL)', Mtuberculosis, re.DOTALL)
if existe:
    for title in existe:
        m = re.match( r'TITLE\s+(.+)', title, re.DOTALL )
        print( re.sub(r'\s+', ' ', m.group(1) ) )
```

```python
#imprimir o locus e a source

flag = False
source = ""
locusid = ""
fonte = open("Mtuberculosis.gb",'r')
for linha in fonte:
    if re.search(r"^LOCUS", linha):
        s = re.match(r"LOCUS\s+([^ ]+)", linha)
        if s:
            locusid = s.group(1)    
    if re.search(r"^SOURCE", linha):
        s = re.match(r"SOURCE\s+(.+)", linha)
        if s:
            source = s.group(1)
    if re.search(r"//", linha):
        flag = False  
    if re.search(r"ORIGIN", linha):
        flag = True
fonte.close()
print( locusid, source)
```

```python
#ler o ficheiro
Mtuberculosis = SeqIO.read("Mtuberculosis.gb", "genbank")

#Bactéria
print("ID: ", Mtuberculosis.id)
print("Descrição: ", Mtuberculosis.description)

#Locus
print("Locus: ", Mtuberculosis.name)
print("bp: ", len(Mtuberculosis))

#Sequência
print("Tamanho da sequência: ", len(Mtuberculosis), "pb")

#Features
print("Features: ", Mtuberculosis.features)
print("Tamanho das features: ", len(Mtuberculosis.features))
#Identificação de features do tipo CDS e a sua localização:
featcds = [] 
for i in range(len(Mtuberculosis.features)):
    if Mtuberculosis.features[i].type=='CDS':
        featcds.append(i)
for k in featcds:
    print("Features do tipo CDS e a sua localização:\n",Mtuberculosis.features[k].location)
    print(len(featcds))

featgenes = []  #número de genes anotados no registo (feature tipo “gene”)
for i in range(len(Mtuberculosis.features)):
    if Mtuberculosis.features[i].type == "gene":
        featgenes.append(i)
print(len(featgenes))

#Reference
print("Referências bibliográficas relacionadas com registo:\n", Mtuberculosis.annotations['references'])

#Anotations
print(len(Mtuberculosis.annotations) )
print("from:", Mtuberculosis.annotations["source"])
print(Mtuberculosis.letter_annotations)
print("Organismo:", Mtuberculosis.annotations["organism"])
print("Taxonomia:", Mtuberculosis.annotations["taxonomy"])
print("Tipo de molécula: ", Mtuberculosis.annotations['molecule_type'])
print("Código do registo: ", Mtuberculosis.annotations['accessions'])
print ("Keywords:\n",Mtuberculosis.annotations.keys())


print(Mtuberculosis.dbxrefs)
```
